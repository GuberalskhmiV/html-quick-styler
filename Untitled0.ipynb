````python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_id = "ibm-granite/granite-3.3-2b-instruct"

tokenizer = AutoTokenizer.from_pretrained(model_id)
model = AutoModelForCausalLM.from_pretrained(
    model_id,
    torch_dtype=torch.float16,
    device_map="auto"
)

def generate_text(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to(model.device)
    output = model.generate(
        **inputs,
        max_new_tokens=500,
        temperature=0.7
    )
    return tokenizer.decode(output[0], skip_special_tokens=True)

print("Model Loaded Successfully ✅")
````

````python
print(generate_text("Write a simple HTML page for a portfolio website"))
````

````python
def build_html_page(description):

    prompt = f"""
    Create a visually stunning modern website using HTML5 and CSS3.

    Requirements:
    - Dark theme
    - Gradient hero
    - Sticky navbar
    - Glass cards
    - Hover animations
    - Responsive layout

    Website description:
    {description}

    Return ONLY pure HTML code.
    Do NOT explain anything.
    Do NOT repeat instructions.
    """

    html = generate_text(prompt)

    # Try to extract only HTML part
    start = html.find("<!DOCTYPE")
    if start == -1:
        start = html.find("<html")

    if start != -1:
        html = html[start:]

    return html
````

````python
import gradio as gr

def generate_page(description):
    if not description.strip():
        return """
        <div style="padding:40px; text-align:center; font-family:sans-serif;">
            <h2 style="color:#ff4b5c;">Please enter a description ✨</h2>
        </div>
        """

    return f"""
    <html>
    <head>
        <style>
            body {{
                margin: 0;
                padding: 0;
                font-family: 'Poppins', sans-serif;
                background: linear-gradient(135deg, #667eea, #764ba2, #ff758c);
                background-size: 400% 400%;
                animation: gradientBG 12s ease infinite;
                color: white;
                text-align: center;
            }}

            @keyframes gradientBG {{
                0% {{ background-position: 0% 50%; }}
                50% {{ background-position: 100% 50%; }}
                100% {{ background-position: 0% 50%; }}
            }}

            .card {{
                margin: 80px auto;
                padding: 40px;
                width: 70%;
                background: rgba(255, 255, 255, 0.15);
                border-radius: 20px;
                backdrop-filter: blur(15px);
                box-shadow: 0 8px 32px rgba(0,0,0,0.3);
            }}

            h1 {{
                font-size: 48px;
                margin-bottom: 20px;
                background: linear-gradient(to right, #ffffff, #ffeaa7);
                -webkit-background-clip: text;
                -webkit-text-fill-color: transparent;
            }}

            p {{
                font-size: 20px;
                line-height: 1.6;
            }}

            .footer {{
                margin-top: 40px;
                font-size: 14px;
                opacity: 0.8;
            }}
        </style>
    </head>
    <body>
        <div class="card">
            <h1>✨ Your Creative Web Page ✨</h1>
            <p>{description}</p>
            <div class="footer">
                Generated with ❤️ using Gradio
            </div>
        </div>
    </body>
    </html>
    """

with gr.Blocks(theme=gr.themes.Soft(primary_hue="purple", secondary_hue="pink")) as app:

    gr.Markdown(
        """
        # 🎨 Creative Web Page Generator
        Turn your ideas into a beautiful colorful web design ✨
        """
    )

    description = gr.Textbox(
        label="Describe your dream web page",
        placeholder="Example: A futuristic AI startup landing page with neon colors...",
        lines=5
    )

    btn = gr.Button("✨ Generate Magic")

    output = gr.HTML()

    btn.click(
        fn=generate_page,
        inputs=description,
        outputs=output
    )

if __name__ == "__main__":
    app.launch()
````

````python
def build_html_page(description):

    prompt = f"""
    You are a professional web designer.

    Create a visually stunning, modern, and professional website using HTML5 and CSS3.

    Design Requirements:
    - Dark theme with gradient highlights
    - Sticky navbar with blur background
    - Large hero section with gradient text
    - Glassmorphism cards
    - Box shadows
    - Smooth transitions
    - Rounded corners
    - Clean typography (use Google Font like Poppins)
    - Mobile responsive
    - Hover animations
    - Button ripple effect

    Website description:
    {description}

    Important:
    - Include full <style> inside HTML
    - Do NOT include markdown
    - Return only HTML code
    """

    html = generate_text(prompt)
    return html.replace("```html", "").replace("```", "")
````
